```python
import pandas as pd

# Confirmed cases
confirmed_url = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv"
confirmed_df = pd.read_csv(confirmed_url)
print("Confirmed cases:\n", confirmed_df.head())

# Deaths
deaths_url = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv"
deaths_df = pd.read_csv(deaths_url)
print("\nDeaths:\n", deaths_df.head())

# Recovered cases
recovered_url = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv"
recovered_df = pd.read_csv(recovered_url)
print("\nRecovered cases:\n", recovered_df.head())
```

```python
deaths_df = deaths_df.drop('Province/State',axis =1)
# Melt the dataframe
id_vars = ['Country/Region', 'Lat', 'Long']
deaths_df_long = deaths_df.melt(id_vars=id_vars, var_name='Date', value_name='Confirmed_death')

# Convert the 'Date' column to datetime format
deaths_df_long['Date'] = pd.to_datetime(deaths_df_long['Date'])


deaths_df_long = deaths_df_long.groupby(['Country/Region','Lat','Long','Date'])['Confirmed_death'].sum().reset_index()
```

```python
confirmed_df = confirmed_df.drop('Province/State',axis =1)
# Melt the dataframe
id_vars = ['Country/Region', 'Lat', 'Long']
confirmed_df_long = confirmed_df.melt(id_vars=id_vars, var_name='Date', value_name='Confirmed')

# Convert the 'Date' column to datetime format
confirmed_df_long['Date'] = pd.to_datetime(confirmed_df_long['Date'])

# Print the first few rows of the long format dataframe
confirmed_df_long = confirmed_df_long.groupby(['Country/Region','Lat','Long','Date'])['Confirmed'].sum().reset_index()
```

```python
confirmed_df_long.loc[confirmed_df_long['Country/Region'] == 'Canada']
```

```python
recovered_df = recovered_df.drop('Province/State',axis =1)
# Melt the dataframe
id_vars = ['Country/Region', 'Lat', 'Long']
recovered_df_long = recovered_df.melt(id_vars=id_vars, var_name='Date', value_name='Recovered')

# Convert the 'Date' column to datetime format
recovered_df_long['Date'] = pd.to_datetime(recovered_df_long['Date'])

# Print the first few rows of the long format dataframe
recovered_df_long = recovered_df_long.groupby(['Country/Region','Lat','Long','Date'])['Recovered'].sum().reset_index()
```

```python
#
full = pd.merge(deaths_df_long,recovered_df_long, how ='left', on = ['Country/Region','Date','Lat','Long'])
```

```python

full = full.merge(confirmed_df_long, how ='left', on = ['Country/Region','Date','Lat','Long'] )
```

```python
# Assuming your DataFrame is named 'full'
country_list = full['Country/Region'].unique()

all_results = []

for country in country_list:
    result = full.loc[full['Country/Region'] == country][['Country/Region', 'Date','Lat','Long','Confirmed','Confirmed_death','Recovered']]
    result['Confirmed_diff'] = result['Confirmed'].diff()
    result['recovered_diff'] = result['Recovered'].diff()
    result['Confirmed_death_diff'] = result['Confirmed_death'].diff()
    result = result.reset_index(drop=True)
    all_results.append(result)

# Concatenate all the results into a single DataFrame
final_result = pd.concat(all_results, ignore_index=True)

final_result = final_result.fillna(0, inplace = False)
```
